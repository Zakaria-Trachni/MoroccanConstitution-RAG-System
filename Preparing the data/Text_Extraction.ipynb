# ***Extract the Text using PyPDF2 Library :***

```python
import PyPDF2 as pdf
import re
import os
```

***Extract the text from the PDF file***

```python
filename = f"..\\Morocco Constitution 2011-English.pdf"

reader = pdf.PdfReader(filename)

Titles_page = reader.pages[1].extract_text()

pages_copy = []
for page in reader.pages:
    pages_copy.append(page.extract_text())

# print(*pages_copy, sep="\n")
```

```python
for j, page in enumerate(pages_copy):
    pages_copy[j] = page.split("Morocco 2011")[-1]
```

***Join the text pages in One Text***

```python
text = "".join(pages_copy[2:-6])
print(text)
```

***Put the text in a texte file***

```python
with open('Unstructured_Topic_indexes_Text.txt', 'w') as file:
    file.write(text)
    file.close()
```

# ***Extract the Text using PyPDF2 Library :***

```python
from pdfminer.high_level import extract_text

text = extract_text(filename)

print(text)
```

***Make the text Lower and remove the upper table content***

```python
# text = text.lower()

# preamble = re.search(r"\npreamble\s\s\(cid:127\)",text)
preamble = re.search(r"\nPreamble\s\s\(cid:127\)",text)
preamble.start()

text = text[preamble.start():]
print(text)
```

***Remove all the lower text of the pages from all the text***

```python

L = list(re.finditer(r"\sMorocco\s2011\s+Page\s\d+\s",text))
liste = []
for i, x in enumerate(L):
    if i == 0:
        liste.append(text[:x.start()])
    else:
        liste.append(text[L[i-1].end():x.start()])

print(len(liste))

text = "\n".join(liste)

print(text)
```

***Remove all the upper text of the pages from all the text***

```python
text1 = text
```

```python
# L = list(re.finditer(r"\sconstituteproject\.org\s\spdf\sgenerated:\s23\snov\s2017,\s15:19\s", text1))
L = list(re.finditer(r"\sconstituteproject\.org\s\sPDF\sgenerated:\s23\sNov\s2017,\s15:19\s", text1))
print(len(L))
liste = []
for i, x in enumerate(L):
    if i == 0:
        liste.append(text1[:x.start()])
    else:
        liste.append(text1[L[i-1].end():x.start()])


text1 = "\n".join(liste)

print(text1)
```

```python
# L = list(re.finditer(r"\sconstituteproject\.org\s\spdf\sgenerated:\s23\snov\s2017,\s15:19\s", text1))
L = list(re.finditer(r"\(cid:127\)( +\w+\W?\w+(\(\w\))?)+\n", text1))
print(len(L))
# for x in L:
    # print(x.group(1))
print(*L, sep="\n")
```

```python
liste = []
for i, x in enumerate(L):
    if i == 0:
        liste.append(text1[:x.start()])
    else:
        liste.append(text1[L[i-1].end():x.start()])

liste.append(text1[L[-1].end():])

# print(*liste, sep="\n")
text1 = "\n".join(liste)

print(text1)
```

***Devide the text According to titles***

```python
text2 = text1
print(text2)
```

***Set up the titles pattern and find all the corresponding titles***

```python
# Titles_pattern = r"\ntitle.+\s"
Titles_pattern = r"\nTitle.+\s"

print(*list(re.finditer(Titles_pattern,text2)), sep="\n")
# for x in re.finditer(Titles_pattern,Titles_page):
#     print(x)
```

***Devide the text into a list of titles and its articles***

```python
L = list(re.finditer(Titles_pattern, text2))
liste = []
for i, x in enumerate(L):
    if i == 0:
        liste.append(text2[:x.start()])
    else:
        liste.append(text2[L[i-1].start():x.start()])

liste.append(text2[L[-1].start():])

print(len(liste))

# print(*liste, sep="\n================================\n================================")
```

***Put subtexts into text files, each with a title***

```python
num = 0

os.makedirs("Titles")

for title in liste:
    if num == 0:
        with open(f"Titles/Preamble.txt", "w") as file:
            file.write(title)
            file.close()
    else:
        os.makedirs(f"Titles/Title {num:02d}")
        with open(f"Titles/Title {num:02d}/Title {num:02d}.txt", "w") as file:
            file.write(title)
            file.close()

    num += 1
```
